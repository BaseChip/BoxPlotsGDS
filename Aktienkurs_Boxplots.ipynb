# Visualisierung des Aktienkurses der Bayer AG als BoxPlots

Der unten aufgeführte Code dient als Beispiel um zu zeigen, wie mithilfe der Python Bibilotheken Matplotlib und Pandas BoxPlots visualisiert werden können.
Dafür werden die Aktienkurs Daten aus der unten angegebenen [Datei](https://raw.githubusercontent.com/BaseChip/BoxPlotsGDS/main/BAYN.DE.csv) extrahiert und der Schlusswert (variable: column) der Aktie der letzten x Tage (variable: days) für n Anzahl an Plots (variable: plots) visualisiert. 


```python
# Importieren der benötigten Bibliotheken
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Pfad zur CSV-Datei
file_path = 'https://raw.githubusercontent.com/BaseChip/BoxPlotsGDS/main/BAYN.DE.csv'

# Anzahl der letzten Werte
days = 3

# Anzahl der Iterationen
plots = 5

# Zu visualisierender Wert
column = "Close"

# CSV-Datei einlesen
df = pd.read_csv(file_path)

# Umwandeln der 'Date' Spalte in datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
# Erstellen einer Figur und mehrerer Achsen (subplots)
# Wir setzen die Figurgröße auf 15x7 (Breite x Höhe)
fig, axs = plt.subplots(1, plots, figsize=(15, 7))

for i in range(plots):
    # Die letzten 'last_n_values' Werte der Spalten 'Close' und 'Date' auswählen
    data_close = df[column].tail(days * (plots - i))
    data_date = df['Date'].tail(days * (plots - i))

    # Nur die ersten 'last_n_values' Datenpunkte auswählen
    data_close = data_close.head(days)
    data_date = data_date.head(days)

    # Erstellen eines Boxplots aus der Serie, Breite der Boxen anpassen
    boxplot = axs[i].boxplot(data_close, widths=0.5)

    # Berechnen des Mittelwerts
    mean = np.mean(data_close)

    # Finden der x-Koordinaten der Box
    x = boxplot['boxes'][0].get_xdata()

    # Hinzufügen der Linie für den Mittelwert
    axs[i].plot([x[0], x[2]], [mean, mean], color='r', linestyle='dashed', linewidth=2)

    # Hinzufügen vom Titel mit dem Zeitraum
    axs[i].set_title(data_date.min().strftime('%d.%m') + ' - ' + data_date.max().strftime('%d.%m.%Y'))

# Anpassen des Layouts, um Überlappungen zu vermeiden
plt.tight_layout()

# Anzeigen der Figur
plt.show()
```

## Darstellung der Schlusswerte in einem Diagramm

```python
# Initialisieren einer Liste für die Daten und einer für die Labels
data_list = []
label_list = []

for i in range(plots):
    # Die letzten 'days' Werte der Spalten 'Close' und 'Date' auswählen
    data_close = df['Close'].tail(days * (plots - i))
    data_date = df['Date'].tail(days * (plots - i))

    # Nur die ersten 'days' Datenpunkte auswählen
    data_close = data_close.head(days)
    data_date = data_date.head(days)

    # Hinzufügen der Daten und Labels zu den jeweiligen Listen
    data_list.append(data_close)
    label_list.append(data_date.min().strftime('%d.%m.%Y') + ' - ' + data_date.max().strftime('%d.%m.%Y'))

# Größe des Plots anpassen
plt.figure(figsize=(12, 8))

# Erstellen eines Boxplots aus der Liste der Daten, Breite der Boxen anpassen
boxplot = plt.boxplot(data_list, widths=0.5, labels=label_list, vert=True)

# Berechnen der Mittelwerte und Hinzufügen der Linien für die Mittelwerte
for i in range(plots):
    mean = np.mean(data_list[i])
    x = boxplot['boxes'][i].get_xdata()
    plt.plot([x[0], x[2]], [mean, mean], color='r', linestyle='dashed', linewidth=1)

# Hinzufügen vom Titel
plt.title(f'Schlusswert des Aktienkurses der Bayer AG für jeweils {days} Tage')

# Rotieren der x-Achsen Labels
plt.xticks(rotation=45)

# Anzeigen des Plots
plt.show()
```

## Visualisierung des Aktienkurses der Bayer AG für einen einzelnen Tag

In dieser Variante wird der Boxplot nicht aus den Schlusswerten generiert sondern aus den Werten Open	High Low und Close

```python
plots = 10
```

```python
# Initialisieren einer Liste für die Daten und einer für die Labels
data_list = []
label_list = []

for i in range(plots):
    # Die letzten 'plots' Werte der Spalten 'Open', 'High', 'Low', 'Close' und 'Date' auswählen
    data_values = df[['Open', 'High', 'Low', 'Close']].tail(plots - i).head(1)
    data_date = df['Date'].tail(plots - i).head(1)

    # Umwandeln der ausgewählten Werte in eine Liste und Hinzufügen zur data_list
    data_list.append(data_values.values[0])
    
    # Hinzufügen des Datums zur label_list
    label_list.append(data_date.iloc[0].strftime('%d.%m.%Y'))

# Größe des Plots anpassen
plt.figure(figsize=(12, 8))

# Erstellen eines Boxplots aus der Liste der Daten, Breite der Boxen anpassen
boxplot = plt.boxplot(data_list, widths=0.5, labels=label_list, vert=True)

# Berechnen der Mittelwerte und Hinzufügen der Linien für die Mittelwerte
for i in range(plots):
    mean = np.mean(data_list[i])
    x = boxplot['boxes'][i].get_xdata()
    plt.plot([x[0], x[2]], [mean, mean], color='r', linestyle='dashed', linewidth=1)

# Hinzufügen vom Titel
plt.title('Aktienkurs der Bayer AG')

# Rotieren der x-Achsen Labels
plt.xticks(rotation=45)

# Anzeigen des Plots
plt.show()
```
